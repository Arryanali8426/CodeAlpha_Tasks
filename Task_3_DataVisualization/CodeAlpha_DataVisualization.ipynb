```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Question: How are math scores distributed among students?
plt.hist(df['Math Score'])
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('StudentsPerformance.csv')
df.head()

df.info()
df.isnull().sum()
df.describe()
df.groupby('Gender')[['Math Score', 'Reading Score', 'Writing Score']].mean()
```

```python
# Question: What is the gender distribution in the dataset?
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
# Question: Do males or females perform better in math?
sns.boxplot(x='Gender', y='Math Score', data=df)
plt.title("Math Score Comparison by Gender")
plt.show()
```

## Data Visualization Summary

In this task, different visualizations were created to better understand the dataset.
Histograms were used to analyze score distributions, bar charts showed gender counts,
and box plots helped compare performance between groups.
Visual representation made patterns and comparisons easier to interpret.
